**Question 1:** Load ``SpotifySongs.csv``. Sort the values by ``release_date`` in ascending order and reset the index. Don't preserve the old values. Print the last 3 rows.

*Hint: To check the type of a pandas Series, add the `dtype` attribute. If you would like to check the types of all the columns in a DataFrame, use the `dtypes` attribute.*

```python
#answer here
songs_df = pd.read_csv("SpotifySongs.csv")

# Sort values by release date in ascending order and reset the index
songs_df = songs_df.sort_values(by="release_date", ascending=True).reset_index(drop=True)

# Print the last 3 rows of the dataframe
songs_df.tail(3)
```

**Question 2:**  Remove any duplicate songs while keeping the last instance. For the purpose of this exercise, we will consider two songs as duplicates if they have the same artists, name and duration.

*Hint: Look up the `drop_duplicates()` pandas function*

```python
#answer here
songs_df = songs_df.drop_duplicates(subset=["name", "artists", "duration_ms"], keep="last")
```

**Question 3:** Modify ``artists`` column, such that the names are separated by a comma and a space, i.e. drop the square brackets and the quotes. For example, ``['Jerry Garcia Band', 'Jerry Garcia']``, should become ``Jerry Garcia Band, Jerry Garcia``.

*Hint: Two ways can be: (1) using `.apply()` with the `re.sub()` function, or (2) using `.replace()`*

```python
#answer here
# **Question 3:** Modify ``artists`` column, such that the names are separated by a comma and a space, i.e. drop the square brackets and the quotes. For example, ``['Jerry Garcia Band', 'Jerry Garcia']``, should become ``Jerry Garcia Band, Jerry Garcia``.

songs_df["artists"] = songs_df["artists"].str.replace("[", "").str.replace("]", "").str.replace("'", "")
```

**Question 4:** Load ``SpotifyTopSongsByCountry.csv`` and drop the ``Duration`` and ``Explicit`` columns. Merge it with the songs dataframe we used in the past questions. Merge it on artists and song title. Note that the column names are different in the dataframes.

```python
#answer here
top_songs_df = pd.read_csv("SpotifyTopSongsByCountry.csv")

# Drop the Duration and Explicit columns
top_songs_df = top_songs_df.drop(columns=["Duration", "Explicit"])

# Rename related columns in top_songs_df and merge with songs_df
top_songs_df = top_songs_df.rename(columns={"Title": "name", "Artists": "artists"})
top_songs_df = top_songs_df.merge(songs_df, on=["name", "artists"])

top_songs_df
```
