**Question 1:**

Create a Pandas Series called ``ID`` with the following values:

```
id = [23, 43, 12, 13, 67, 89, 90, 56, 34]
```

Then create a Pandas dataframe called ``students`` from the given dictionary.

```
dct = {'Name': {0: 'Sam', 1: 'Dave', 
                2: 'Yasmeen', 3: 'Ahmed', 
                4: 'Arjun', 5: 'Salma', 
                6: 'Michael', 7: 'Chalsea', 
                8: 'Simone'}, 
         
       'Marks': {0: None, 1: 97, 2: 45, 3: 78, 
                 4: 56, 5: 76, 6: 100, 7: 87, 
                 8: 89}, 
         
       'Grade': {0: 'B', 1: 'A', 2: 'F', 3: 'C', 
                 4: 'E', 5: 'C', 6: 'A', 7: 'B', 
                 8: 'B'} 
       } 
```

1. Add ``ID`` to the dataframe.
2. Change Yasmeen's grade to ``D``. (*Bonus*: Can you come up with at least 2 different formats to change the Yasmeen's grade?)
3. Make the column ``ID`` the index.
4. Someone removed Sam's mark. Replace it with the mean of Chalsea and Simon's Marks who also got ``B``?
5. Increase the students' marks by 5, except for those who got A.

````python
# answer here
id = [23, 43, 12, 13, 67, 89, 90, 56, 34]

dct = {'Name': {0: 'Sam', 1: 'Dave',
                2: 'Yasmeen', 3: 'Ahmed',
                4: 'Arjun', 5: 'Salma',
                6: 'Michael', 7: 'Chalsea',
                8: 'Simone'},

       'Marks': {0: None, 1: 97, 2: 45, 3: 78,
                 4: 56, 5: 76, 6: 100, 7: 87,
                 8: 89},

       'Grade': {0: 'B', 1: 'A', 2: 'F', 3: 'C',
                 4: 'E', 5: 'C', 6: 'A', 7: 'B',
                 8: 'B'}
       }

# Create a dataframe from the dictionary
df = pd.DataFrame(dct)
df
````

````python
# 1. Add ID column to the dataframe
df['ID'] = id
df
````

````python
# 2. Change Yasmeen's grade to D
df.loc[df['Name'] == 'Yasmeen', 'Grade'] = 'D'

# Alternate method:
df.at[2, 'Grade'] = 'D'

df
````

````python
# 3. Make the column ID the index of the dataframe
df.set_index('ID', inplace=True)
df
````

````python
# 4. Sam's marks = average of Simone's and Chalsea's marks
grade = (df.loc[df['Name'] == 'Simone', 'Marks'].values[0] + df.loc[df['Name'] == 'Chalsea', 'Marks'].values[0]) / 2
df.loc[df['Name'] == 'Sam', 'Marks'] = grade
df
````

````python
# 5. If student's grade is not A, increase their marks by 5
df.loc[df['Grade'] != 'A', 'Marks'] += 5
df
````

**Question 2:** Load the camera dataset csv file into a pandas DataFrame and do the following:

1. Drop all columns except "Model", "Release date", "Weight (inc. batteries)", "Storage included", "Price".
2. Rename "Release date" to "Year", "Weight (inc. batteries)" to "Weight" and "Storage included" to "Storage"
3. Extract all camera models that were produced after 2003 and has weight and storage. 
4. What's average price for cameras that have storage greater than or equal to the median storage?


*Hint: For the first task, look up .isin() and ~ (not).*

````python
#answer here
path = 'data/camera_dataset.csv'
````

````python
df = pd.read_csv(path)
df
````

````python
# 1. Drop all columns except Model, Release date, Weight (inc. batteries), Storage included, Price
df = df[['Model', 'Release date', 'Weight (inc. batteries)', 'Storage included', 'Price']]
df
````

````python
# 2. Rename "Release date" to "Year", "Weight (inc. batteries)" to "Weight" and "Storage included" to "Storage"
df.rename(columns={'Release date': 'Year', 'Weight (inc. batteries)': 'Weight', 'Storage included': 'Storage'}, inplace=True)
df
````

````python
# 3. Extract all camera models that were produced after 2003 and has weight and storage.
df = df[(df['Year'] > 2003) & (df['Weight'].notnull()) & (df['Storage'].notnull())]
df
````

````python
# 4. What's average price for cameras that have storage greater than or equal to the median storage?
median = df['Storage'].median()
df.loc[df['Storage'] >= median, 'Price'].mean()
````
