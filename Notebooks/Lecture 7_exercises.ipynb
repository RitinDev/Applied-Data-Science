**Question 1:** Define a function `isPIN`, when given a string, checks if it is a valid ATM PIN. It should be either 4 or 6 digits, no characters, and cannot start with 0.

*Hint*: Check out `re.fullmatch()` in the [regex document](https://docs.python.org/3/library/re.html)

````python
def isPIN(s):
  # Validate s: It should be either 4 or 6 digits long, no characters other than digits and no leading zeros (if it is 6 digits long)
  if re.match(r"^[1-9][0-9]{3}$", s) or re.match(r"^[1-9][0-9]{5}$", s):
    return True
  return False


print(isPIN("153333")) #True
print(isPIN("053333")) #False
print(isPIN("1533")) #True
print(isPIN("15333")) #False
print(isPIN("153333 is a pin")) #False
````

**Question 2:** Are these two letters always friends? Write a function `friends` that will take as input parameters a sentence and two letters. The function will output `True` if the second letter *always* appears after the first one, `False` otherwise.

````python
def friends(s, l1, l2):
    # Check if l2 always follows l1 in s
    return len(re.findall(l1+l2, s)) == len(re.findall(l1, s))


print(friends("the sound sounds good", "s", "o")) # False
print(friends("the key key is this", "k", "e")) # True
````

**Question 3:** Create a regex pattern that will take the text below and return the following dictionary:

```
{'animal': 'horses', 'verb': 'are', 'adjective': 'fast'}
{'animal': 'turtles', 'verb': 'are', 'adjective': 'slow'}
{'animal': 'slothes', 'verb': 'are', 'adjective': 'lazy'}
```

````python
text = 'horses are fast, turtles are slow, slothes are lazy.'

#answer here
pattern = """
(?P<animal>\w+) # the animal name
\s
(?P<verb>\w+) # the verb name
\s
(?P<adjective>\w+) # the adjective name
"""

for sentence in re.finditer(pattern, text, re.VERBOSE):
    print(sentence.groupdict())
````

**Question 4:** Extract all names in the text that start with an `A` and end with an `a`.

````python
names = "Ava, Barb, Alyssa, Alex, Tom, Alexandra, John, Alicia"

#answer here
pattern = r"\bA[a-z]*a\b"
print(re.findall(pattern, names))
````

**Question 5:** Given the tagged HTML text below, use regex to generate the following dictionary.

```
{'title': 'Teaching Python', 'author': 'Bedoor AlShebli', 'chapter_title': ''}
{'title': 'Python 101', 'author': 'Anahit Sargsyan', 'chapter_title': 'Chapter 1'}
```

````python
html = """<title>Teaching Python</title> some text <author>Bedoor AlShebli</author> <chapter_title></chapter_title> there is a missing chapter title
          some random text
          ...
          some more random text
          <title>Python 101</title>, <author>Anahit Sargsyan</author>, <chapter_title>Chapter 1</chapter_title>
       """

#answer here
pattern = """
<title>(?P<title>.+?)</title> # the title name
.*?
<author>(?P<author>.+?)</author> # the author name
.*?
<chapter_title>(?P<chapter_title>.*)</chapter_title> # the chapter title name
"""

for sentence in re.finditer(pattern, html, re.VERBOSE):
      print(sentence.groupdict())
````

**Question 6:** Find all numbers which are not separated from the front and back word by a white space. Just provide the pattern string. If the correct regex pattern is provided, the output should be:

```
['010292291', '1414757474']
```
or
```
010292291
1414757474
```

````python
text = "my phone number is010292291this is a cool phone number: 314295291 Yeah1414757474this is another phone number."

#answer here
# Find all phone numbers in the text NOT surrounded by white spaces
# The number should be 9 or 10 digits long
pattern = r"(?<!\s)\d{9,10}(?!\s)"
print(re.findall(pattern, text))
````
