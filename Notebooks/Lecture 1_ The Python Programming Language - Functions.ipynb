**Question 1b:**

Now, let's add an argument $metric$ to the function, that will take either C or F. The conversion should be based on this argument, i.e. if the value being passed is "F" then the function converts to Celsius. 

*Note: the formula for Farenheit to Celsius is $(32°F − 32) × 5/9 = 0°C$*

````python
#answer here
def convert(x: int, metric: chr):
    if metric == "C" or metric == "c":
        return (x * 9 / 5) + 32
    elif metric == "F" or metric == "f":
        return (x - 32) * 5 / 9
    else:
        return "Wrong metric"

print(convert(0, "C"))
print(convert(32, "F"))
print(convert(32, "C"))
````

## Conditionals:

Conditionals give you the power to execute some code if a condition is met or in other words if a condition is "`True`". 

To understand how, we need to first understand the **Boolean Expression**.


### Boolean Expressions

A boolean expression is an expression that evaluates to `True` or `False`.

Example of boolean expressions in the english language are: 
1. It is raining. 
2. Python is a scripting language.

Below we explore some boolean expressions in python.

####Comparison operators

Comparison operators give you a boolean result as a result of comparison.


````python
x = 1
y = 2

result1 = (x > y)  # Parentheses make it easier to read this statement...
result2 = y > x  # ... but they arent compulsory to get your code to work
result3 = x == y  #since a single = is used for variable assignments, == is used to test equality
result4 = x >= y
result5 = x <= y

#print() is one of many built-in functions found in python
#print() allows you to display output anywhere in the code, not just the last line
print(result1, result2, result3, result4, result5)
````

````python
# Also works for strings!
str1 = "abc"
str2 = "abcd"
str3 = "abc"

result1 = str1 == str2
result2 = str1 == str3

print(result1, result2)
````

####Chaining booleans / Logical operators

You may have situations where you'd like see if multiple conditions are true. 

To enable you to do so, python gives you 3 operators: `and`, `or` & `not`.These three words are "keywords" in python.

Look at the examples below, and refer to this "Truth Table" to understand how different booleans interact with each other:

A | B | A `and` B | A `or` B | `not` A
--- | --- | --- | --- | ---
**`False`** | **`False`** | `False` | `False` | `True`
**`False`** | **`True`** | `False` | `True` | `True`
**`True`** | **`False`** | `False` | `True` | `False`
**`True`** | **`True`** | `True` | `True` | `False`

What will the output of each result be?

````python
x = 1
y = -1

# Check if x & y are both positive
result1 = (y > 0) and (x > 0)
print("result1", result1)

# Check if x & y are both on the same side of the origin
result2 = ((y > 0) and (x > 0)) or ((y < 0) and (x < 0))  # The parenthesis determine order of evaluation

print("result2", result2)

x = -1
y = -1

# Check if x & y are both positive
result3 = (y > 0) and (x > 0)
print("result3", result3)

# Check if x & y are both on the same side of the origin
result4 = ((y > 0) and (x > 0)) or ((y < 0) and (x < 0))
print("result4", result4)

# Check if x & y are both negative (using `not` operator)
result5 = (not (y > 0)) and (not (x > 0))
print("result5", result5)
````

### if-else statements

Now that we know about boolean expressions, lets apply them to conditionals!

An if-else statement simply ensures a certain codeblock is run only if its corresponding condition is met. It's syntax is:



```
if <condition>:
  <code block 1>
elif <condition>:
  <code block 2>
elif <condition>:
  <code block 3>
.....
.....
else:
  <code block n>
```

Some technicalities:
1. The `if` is compulsory, and you can have only one `if` in an if-else statement.
2. The `elif` isn't compulsory and you can have any number of `elif`
 conditions. `elif` is short for *else if*.
3. The `else` isn't compulsory and you can have only one `else` statement.
4. Another thing to note here is the presence of `:` and the indendation. The indentation is required. Without it, you will see an error. In most cases you will see that Google Colab automatically indents statements after a `:` for you.

Let's look at some examples!

When talking to someone, you may say something like:
If it is raining, take an umbrella

This basically means:
`If (raining == True)` execute action `take_umbrella`

The concept is similar when it comes to programming.


````python
is_raining = True

if is_raining:  # since condition 'is_raining' evaluates to True, we will enter this code block
    print("Take an umbrella")
````

````python
is_raining = False

if is_raining:
    print("Take an umbrella")

# This code does nothing since 'is_raining' equates to False.
````

````python
# The 'elif' statement allows you to add as many conditions as you'd like

is_raining = False
is_sunny = True

if is_raining:
    print("Take an umbrella")
elif is_sunny:
    print("Take a cap")
````

````python
# The 'elif' statement allows you to add as many conditions as you'd like

is_raining = True
is_sunny = True

if is_raining:
    print("Take an umbrella")
elif is_sunny:
    print("Take a cap")

# Hmm... is the output what you expected?
````

The above example shows the importance of ordering as well as the exclusivity of conditions. 

Some observations:

1. If we had switched the order of the two conditions like:

```
if is_sunny:
  print("Take a cap")
elif is_raining:
  print("Take an umbrella")
```
We would've gotten `Take a cap` as the output

2. `if` statements only allow any one of the codeblocks to run.



````python
# Lets try chaining some conditions together

is_raining = True
is_sunny = True
if is_raining or is_sunny:
    print("1. Take an umbrella")

is_raining = True
is_sunny = False
if is_raining or is_sunny:
    print("2. Take an umbrella")

is_raining = False
is_sunny = True
if is_raining or is_sunny:
    print("3. Take an umbrella")

is_raining = False
is_sunny = False
if (
    is_raining or is_sunny
):  # This doesn't print anything as both these conditions are False!
    print("4. Take an umbrella")
````

3. The `else` statement is a catchall. The else codeblock executes if none of the previous conditions are met. This is also why else doesn't take in a condition.

````python
is_raining = False
is_sunny = False

if is_raining:
    print("Take an umbrella")
elif is_sunny:
    print("Take a cap")
else:
    print("Go empty handed!")
````

## Loops

Loops are statements that allow you to execute the same piece of code over and over again. Repetitive execution of code is also known as **iteration**.
It is an extremely powerful tool in the world of programming.

There are two main kinds of loops:

1. `for` loops
2. `while` loops




### For loops

For loops in python iterate over items in a sequence (eg. list, string).

The syntax of a for loop is:

```
for <variable(s)> in <sequence>:
    <do stuff>
```

In every iteration of the loop, `<variable>` will take the value of an element of the `<sequence>`. The `for loop` will execute until it is out of elements.

For loops perform something known as **definite iteration**. It simply means we specify the number of repetitions in advance. Over here the number of iterations will be the length of our chosen sequence.

Let's try some examples!

````python
#this create a list of letters - we will learn about lists soon!
seq1 = ["a", "b", "c", "d", "e"]

# Simple loop that prints all the elements of seq1
for letter in seq1:
    print(letter)
````

As we see above, the variable `letter` gets assigned an element of the sequence / list `seq1`(conserving the order of the list) in every iteration. Again, just like `if-else` statements, make note of the indentation. In every iteration, only the indented codeblock will run.

````python
# Here's a simple code snippet to print out the square of all the numbers in a list

nums = [1, 2, 3, 4, 5]

for i in nums:
    print(i * i)
````

####range() function

The `range()` function allows us to create arithmetic sequences by specifying rules. Given that sequences are the backbone of pythonic `for` loops, you will often see the `range()` function used in for loops.

A `range()` takes in 3 arguments as follows:

```
range(start_index, stop_index, step)
```

`stop_index` is the only required argument, the other 2 are optional.

`start_index` has a default value of `0` when not assigned otherwise.

`step` is the distance you want to have between your items.

Note that the range includes the `start_index` in its range, but excludes the `stop_index`.

For instance,

with
```start_index=1, stop_index=7, step=2```

we will see an output:

```[1,3,5]```

````python
#this range only includes one number, the compulsory stop_index. 
#It will create a sequence of numbers from 0 to 4 (remember stop_index value is excluded)
for i in range(5):
  print(i)
````

````python
#let's try it with a step value
#This will create a sequence of odd numbers between 1 and 10.
for i in range(1,10,2):
  print(i)
````

### While Loops

`while` loops, similar to `for` loops, repeatedly iterate through code blocks. The biggest difference is `while` loops end when a pre-specified condition is met. 

The syntax is as follows:

```
while <condition>:
  <do stuff> 
```

The loop keeps going as long as `condition` (which is a boolean expression) evaluates to `True`.

Since there is no guarantee that the condition will always be met, `while` loops perform **indefinite iteration**.

Let's try out some examples!

````python
# A simple while loop

i = 0
while i < 5:
    print(i)
    i = i + 1

# When the value of i becomes 5, we don't re-enter the loop as (i < 5) evaluates to False.
````

As mentioned above, there isn't a guarantee that we converge towards our terminating condition in while loops. When we don't converge, we cause something known, and quite rightly so, as an **infinite loop**. 

We most definitely want to avoid this. Let's try to write a simple infinite loop.

````python
while True:
    print("This will go on forever")

# Run this piece of code at your own risk!
# Hint: The stop button on the top left of the cell might come in handy :)
````

### Special mentions: break and continue

`break` and `continue` are keywords in python that allow you to manipulate the control flow of a loop. 

`break` statement allows you to exit out of a loop when a specific condition is triggered.

````python
print("Trying with valid input")

numbers = [31, 12, 3, 24, 15, 982]

for num in numbers:
    if num < 0:  # we don't enter this code block as all numbers are positive
        print("This for loop expects positive numbers! Exiting.")
        break
    else:
        print(num)

print()
print("Trying with invalid input")

numbers = [31, 12, 3, -24, 15, 982]

for num in numbers:
    if num < 0:  
        print("This for loop expects positive numbers! Exiting.")
        break #the loop stops after encountering negative -24
    else:
        print(num)
````

`continue` keyword allows us skip a part of a specific iteration and continue with the loop.

````python
print("Trying with valid input")

numbers = [31, 12, 3, 24, 15, 982]

for num in numbers:
    if num < 0:  # we don't enter this code block as all numbers are positive
        print("This for loop expects only positive numbers! Skipping.")
        continue
    else:
        print(num)

print()
print("Trying with invalid input")

numbers = [31, 12, 3, -24, 15, 982]

for num in numbers:
    if num < 0:  
        print("This for loop expects only positive numbers! Skipping.")
        continue #the for loop skips the rest of the steps for this iteration and continues to the next element in the list
    else:
        print(num)
````

###Nested Loops

Finally, loops can be used within other loops as well. Let's look at an example. Say you want to print every pairwise combination of student and subject from two lists.

````python
students = ['John','Sarah','Ahmad','Ana']
subjects = ['Math','English','Science','Art','History']

for student in students:
  for subject in subjects:
    print(student, subject)
````

Note that you can infinitely nest for loops within for loops, while loops within while loops and mix it up. However, if you find yourself nesting too much, there is probably a more elegant way to decouple the code and reduce its complexity.
