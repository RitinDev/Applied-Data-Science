## A. Difference in total bill for male and female (25 points)

We would like to first look at how males and females differ in terms of `total_bill` across different days. We would like to be able to compare male and female payers in terms of the *median bill*, the *minimum*, the *maximum*, and the *interquartile range* of the `total_bill` for each `day`.

We would also like you to:

- annotate the median value explicitly for each `sex` for each `day` rounded to 1 decimal point for better visibility and comparison.
- remove the frame of the legend
- change the color palette and font to something else other than the default

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the aesthetic style of the plots
sns.set_style("whitegrid")
sns.set_palette("pastel")
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'

# Create the boxplot
plt.figure(figsize=(10, 6))
box_plot = sns.boxplot(
    x="day", 
    y="total_bill",
    hue="sex",
    data=tips,
    palette="flare",
    fill=False,
    gap=0.1,)

# Calculate medians and annotate
medians = tips.groupby(['day', 'sex'])['total_bill'].median()
rounded_medians= tips.groupby(['day', 'sex'])['total_bill'].median().round(1)
vertical_offset = tips['total_bill'].median() * 0.05  # slight offset from median for display

for xtick in box_plot.get_xticks():
    for i, sex in enumerate(['Male', 'Female']):
        median_val = rounded_medians[box_plot.get_xticklabels()[xtick].get_text(), sex]
        # Appropriately aligning the text
        if i == 0: # For male median labels
            box_plot.text(xtick-0.275, median_val + vertical_offset, median_val,
                horizontalalignment='left', color='black', weight='semibold')
        else: # For female median labels
            box_plot.text(xtick+i*0.275, median_val + vertical_offset, median_val,
                    horizontalalignment='right', color='black', weight='semibold')

# Remove legend frame
box_plot.legend(frameon=False)

# Display the plot
plt.show()
```

```python
# Initializing index for accessing medians later
index = 0
# Going through each x-tick i.e. day
for day_tick in box_plot.get_xticks():
    # Let us now set the x position for the text for male and female by a difference of .2
    x_pos_male_median, x_pos_female_median = day_tick-.2, day_tick+.2
    # Let us now set the y position for the text for male and female by a different of 2
    y_pos_male_median, y_pos_female_median= medians[index] + 2, medians[index+1] + 2
    # Let us now retrieve the actual rounded median value we want to annotate
    median_rounded_male, median_rounded_female = rounded_medians[index], rounded_medians[index+1]
    # Finally setting the box_plot text
    box_plot.text(x_pos_male_median,y_pos_male_median, median_rounded_male,
            horizontalalignment='center',size='small',color='black')
    box_plot.text(x_pos_female_median,y_pos_female_median, median_rounded_female,
            horizontalalignment='center',size='small',color='black')
    # Finally incrementing index
    index +=2
```

```python
# Let us read the happiness dataset
import pandas as pd

df = pd.read_csv("happiness.csv")
df = df[['Ladder score', 'Social support', 'Healthy life expectancy',
         'Freedom to make life choices', 'Generosity',
         'Perceptions of corruption']]

corrs = df.corr()

corrs

# Write your implementation of the function below this line

######### SOLUTION #########
# Plot the heatmap
plt.figure(figsize=(6, 4))
heatmap = sns.heatmap(
    corrs, # Plot correlation matrix
    vmin=-1, vmax=1, # Define limits for correlation values
    annot=True, # Annotate each cell
    cmap='BrBG', # Define color map
    cbar=False, # Remove color bar
    linewidths=10 # Set width of cell borders
    )

# Adjust title styling
heatmap.set_title('Correlation Analysis', fontdict={'fontsize':16 }, pad=12)

# Adjust the tick labels
x_tick_labels = [label.get_text() + ' -' for label in heatmap.get_xticklabels()]
heatmap.set_xticklabels(x_tick_labels, fontsize=12)

y_tick_labels = [label.get_text() + ' -' for label in heatmap.get_yticklabels()]
heatmap.set_yticklabels(y_tick_labels, fontsize=12)

# Draw the heatmap
plt.show()
######### SOLUTION END #########
```
