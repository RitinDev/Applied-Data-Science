## Part III: Adding New Columns (50 points)

Based on the rubric above, we want you to add a new column to the DataFrame called *Overall Percentage* which represents the overall percentage a student has achieved in the course based on the rubric given above.

Note: Your solution should not be more than 6-8 lines of code.

```python
# Write you solution here
############# SOLUTION ###############
grades_df["Overall Percentage"] = (
    grades_df.loc[:, 'P1':'P10'].sum(axis=1) / 1200 * 60  # Sum of top 10 Problem Sets
    + grades_df.loc[:, 'R1':'R10'].sum(axis=1) / 500 * 10  # Sum of Recitations
    + grades_df["Class Attendance"] / 28 * 5  # Class Attendance
    + grades_df["Project"] / 100 * 25  # Project
)

grades_df

############# SOLUTION END ###############
```
