DFSC is a notorious player at planting and propagating organized conspiracies in Canada by uploading fake datasets online. We have gotten hold of the data, and would like to visualize it to understand the narrative they are secretly planning to propagate to possibly call them out.

```python
# Write your implementation below this line

######### SOLUTION #########
# Load dataset via pandas
import pandas as pd
import matplotlib.pyplot as plt

fake_stats_df = pd.read_csv('fake_statistics_canada.csv') # Please adjust path if necessary

# Plotting
fig, ax = plt.subplots(figsize=(14,8))

# Pastel color palette
colors = ['#86CAB7', '#B8D89F', '#FFB28A']

# Plot data
bars = fake_stats_df.set_index('province').plot(kind='bar', ax=ax, logy=True, color=colors)

# Note: I used logy=True to make the plot more readable, since the values for some provinces are much lower than others

# Title and labels
ax.set_title('Comparison of Heart Disease, Cancer, and Vaccine Cases Across Provinces')
ax.set_xlabel('Province')
ax.set_ylabel('Number of Cases')
ax.legend(title='Metric')

# Uncomment the following code to annotate the bars with the exact numbers
# I have commented this code out as a personal preference, as I find it makes the plot cluttered
# for container in bars.containers:
#     ax.bar_label(container, label_type='edge', padding=3, fontsize=8)

# Display the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

######### SOLUTION END #########
```

### Note

I have used a logarthmic scale for the visualization since values for some provinces are very low compared to others, which makes it difficult to visualize the data on a linear scale since the bars for some provinces will be too small to be visible. This is how the graph looks without the log-scale Y-axis:

```python
# Plotting
fig, ax = plt.subplots(figsize=(14,8))
fake_stats_df.set_index('province').plot(kind='bar', ax=ax, color=colors) # no logY parameter

# Title and labels
ax.set_title('Comparison of Heart Disease, Cancer, and Vaccine Cases Across Provinces')
ax.set_xlabel('Province')
ax.set_ylabel('Number of Cases')
ax.legend(title='Metric')

# Display the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Part III: Performance over time (25 points)

It is the annual admissions cycle, and the NYUAD graduate committee is looking at applications for selecting students for their graduate program. There are only two spots left to fill, and four potential candidates. To make a better decision, the graduate committee would like to look at their GPAs as a function of time (over 8 semesters). This will allow them to see who made the most progress in college over time, and which semester they struggled the most and why.

We have provided below a DataFrame that maps each student to their semesterly GPAs over 8 semesters. Create a visualization to help the admission committee get a sense of the performances of the four students over time.

Questions you should ask yourself while doing this part:
1. What is the right kind of visual/graph to juxtapose the performances of the four students over time?
2. What kind of design elements are important for the audience to easily grasp the information being presented?

```python
import pandas as pd
# The GPA dictionary mapping students to their GPAs per semester

gpa_dict = {"John Doe": [3.90, 3.86, 3.0, 3.20, 3.33, 3.50, 3.77, 4.0],
 "Jane Roe": [4.0, 4.0, 3.92, 4.0, 3.88, 4.0, 4.0, 4.0],
 "Richard Roe": [4.0, 4.0, 3.87, 3.2, 3.0, 3.0, 2.68, 2.50],
 "Janie Doe": [2.0, 2.30, 2.45, 2.68, 3.0, 3.25, 3.50, 3.92]
 }

# Columns corresponding to the semesters/terms
cols = ["Fall'16", "Spring'17", "Fall'17", "Spring'18", "Fall'18", "Spring'19", "Fall'19", "Spring'20"]

# Creating a dataframe
df = pd.DataFrame.from_dict(gpa_dict, orient="index", columns=cols)
# Defining index
df.index.name = "student"
df
```

I believe that a line chart will be a good visualization to illustrate how student GPAs have changed across semesters. I make this choice since line charts are often a good choice for comparisons over time.

```python
# Write your implementation below this line

######### SOLUTION #########
# Create a line plot to juxtapose the GPAs of the students over the semesters
fig, ax = plt.subplots(figsize=(10,6))

# Plot data
df.T.plot(kind='line', ax=ax, colormap='Paired')

# Title and labels
ax.set_title('Comparison of Student GPAs Over Time')
ax.set_xlabel('Semester')
ax.set_ylabel('GPA')
ax.legend(title='Student')
######### SOLUTION END #########
```
